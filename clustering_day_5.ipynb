```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
digit=load_digits()
```

```python
df=pd.DataFrame(digit.data)
df['Target'] = digit.target
print(df.head())
```

```python
kMeans=KMeans(n_clusters=10,random_state=42)
kMeans.fit(digit.data)
```

```python
df['clusters']=kMeans.labels_
print(df.head())
```

```python
pca=PCA(n_components=2)
pca_result=pca.fit_transform(digit.data)
```

```python
plt.scatter(pca_result[:,0],pca_result[:,1],c=df['clusters'])
```

```python
cluster_centers = kMeans.cluster_centers_.reshape(10, 8, 8)
fig, axes = plt.subplots(2, 5, figsize=(8, 4))

for i, ax in enumerate(axes.flat):
    ax.imshow(cluster_centers[i], cmap='gray')
    ax.set_title(f'cluster {i}')
    ax.axis('off')

plt.show()
```
